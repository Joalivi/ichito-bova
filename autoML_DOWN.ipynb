```python
data = data.drop(['Unnamed: 0'], axis=1).drop(0)
```

```python
data = data[['bullishHorn', 'bearHorn', 'bullishHigh', 'bearHigh', 'bullishLow',
       'bearLow', 'bullishHarami', 'bearHarami', 'is_EMA8_above_EMA24',
       'is_EMA8_above_EMA48', 'is_EMA24_above_EMA48', 'is_price_above_EMA8',
       'is_price_above_EMA24', 'is_price_above_EMA48', 'target_down']]
```

```python
from pycaret.classification import *
```

```python
setup = setup(data = data, target = 'target_down',
            normalize = True,
            transformation = True,
            pca = True,
            pca_components = 0.95,
            remove_outliers = True,
            #remove_multicollinearity = True,
            create_clusters = True,
            #polynomial_features = True,
            #feature_selection = True
            )
```

```python
compare_models()
```
