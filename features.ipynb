```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
df = pd.read_csv('./databases/VALE3 Historical Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date')
```

```python
df['Price-1'] = df['Price'].shift()
df['High-1'] = df['High'].shift()
df['Low-1'] = df['Low'].shift()
df['Open-1'] = df['Open'].shift()
df = df.drop(0)
```

```python
df = df.drop(['Vol.', 'Change %'], axis=1)
```

```python
df.head()
```

```python
df.tail()
```

```python
#The Eight-Trigram

df['bullishHorn'] = df.apply(lambda x: 1 if ((x['High'] > x['High-1']) &
                                                 (x['Low'] < x['Low-1']) &
                                                 (x['Price'] > x['Price-1'])) 
                                            else 0, axis=1)

df['bearHorn'] = df.apply(lambda x: 1 if ((x['High'] > x['High-1']) &
                                                 (x['Low'] < x['Low-1']) &
                                                 (x['Price'] < x['Price-1'])) 
                                            else 0, axis=1)

df['bullishHigh'] = df.apply(lambda x: 1 if ((x['High'] > x['High-1']) &
                                                 (x['Low'] > x['Low-1']) &
                                                 (x['Price'] > x['Price-1'])) 
                                            else 0, axis=1)

df['bearHigh'] = df.apply(lambda x: 1 if ((x['High'] > x['High-1']) &
                                                 (x['Low'] > x['Low-1']) &
                                                 (x['Price'] < x['Price-1'])) 
                                            else 0, axis=1)

df['bullishLow'] = df.apply(lambda x: 1 if ((x['High'] < x['High-1']) &
                                                 (x['Low'] < x['Low-1']) &
                                                 (x['Price'] > x['Price-1'])) 
                                            else 0, axis=1)

df['bearLow'] = df.apply(lambda x: 1 if ((x['High'] < x['High-1']) &
                                                 (x['Low'] < x['Low-1']) &
                                                 (x['Price'] < x['Price-1'])) 
                                            else 0, axis=1)

df['bullishHarami'] = df.apply(lambda x: 1 if ((x['High'] < x['High-1']) &
                                                 (x['Low'] > x['Low-1']) &
                                                 (x['Price'] > x['Price-1'])) 
                                            else 0, axis=1)

df['bearHarami'] = df.apply(lambda x: 1 if ((x['High'] < x['High-1']) &
                                                 (x['Low'] > x['Low-1']) &
                                                 (x['Price'] < x['Price-1'])) 
                                            else 0, axis=1)
```

```python
df.head()
```

```python
df['EMA8'] = df['Price'].ewm(span=8, adjust=False).mean()
df['EMA24'] = df['Price'].ewm(span=24, adjust=False).mean()
df['EMA48'] = df['Price'].ewm(span=48, adjust=False).mean()
```

```python
#Boolean Markers

#EMAs
df['is_EMA8_above_EMA24'] = df.apply(lambda x:1 if x['EMA8'] > x['EMA24'] else 0, axis=1)
df['is_EMA8_above_EMA48'] = df.apply(lambda x:1 if x['EMA8'] > x['EMA48'] else 0, axis=1)
df['is_EMA24_above_EMA48'] = df.apply(lambda x:1 if x['EMA24'] > x['EMA48'] else 0, axis=1)

#Price
df['is_price_above_EMA8'] = df.apply(lambda x:1 if x['Price'] > x['EMA8'] else 0, axis=1)
df['is_price_above_EMA24'] = df.apply(lambda x:1 if x['Price'] > x['EMA24'] else 0, axis=1)
df['is_price_above_EMA48'] = df.apply(lambda x:1 if x['Price'] > x['EMA48'] else 0, axis=1)

#Distance Price/EMA Percent

df['%Distance_price_EMA8'] = abs((df['Price'] - df['EMA8']) * 100 / (df['EMA8']))
df['%Distance_price_EMA24'] = abs((df['Price'] - df['EMA24']) * 100 / (df['EMA24']))
df['%Distance_price_EMA48'] = abs((df['Price'] - df['EMA48']) * 100 / (df['EMA48']))

#Difference EMA-price

df['Dif_price_EMA8'] = df['Price'] - df['EMA8']
df['Dif_price_EMA24'] = df['Price'] - df['EMA24']
df['Dif_price_EMA48'] = df['Price'] - df['EMA48']
```

```python
df.head()
```

```python
df['target'] = df.apply(lambda x: 1 if (x['High'] >= x['Open']*1.01) else 0, axis=1)
```

```python
df['target'].value_counts()
```

```python
df1 = df[['bullishHorn', 'bearHorn', 'bullishHigh', 'bearHigh',
       'bullishLow', 'bearLow', 'bullishHarami', 'bearHarami', 'is_EMA8_above_EMA24', 'is_EMA8_above_EMA48',
       'is_EMA24_above_EMA48', 'is_price_above_EMA8', 'is_price_above_EMA24',
       'is_price_above_EMA48', '%Distance_price_EMA8', '%Distance_price_EMA24',
       '%Distance_price_EMA48', 'Dif_price_EMA8', 'Dif_price_EMA24',
       'Dif_price_EMA48', 'target']].copy()
```
