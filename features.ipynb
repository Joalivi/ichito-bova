```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
vale = pd.read_csv('./databases/VALE3 Historical Data.csv')
```

```python
vale.head()
```

```python
vale.info()
```

```python
vale['Date'] = pd.to_datetime(vale['Date'])
```

```python
vale['Price-1'] = vale['Price'].shift()
vale['High-1'] = vale['High'].shift()
vale['Low-1'] = vale['Low'].shift()
vale['Open-1'] = vale['Open'].shift()
vale = vale.drop(0)
```

```python
vale = vale.drop(['Vol.', 'Change %'], axis=1)
```

```python
vale.head()
```

```python
vale.tail()
```

```python
#The Eight-Trigram

vale['bullishHorn'] = vale.apply(lambda x: 1 if ((x['High'] > x['High-1']) &
                                                 (x['Low'] < x['Low-1']) &
                                                 (x['Price'] > x['Price-1'])) 
                                            else 0, axis=1)

vale['bearHorn'] = vale.apply(lambda x: 1 if ((x['High'] > x['High-1']) &
                                                 (x['Low'] < x['Low-1']) &
                                                 (x['Price'] < x['Price-1'])) 
                                            else 0, axis=1)

vale['bullishHigh'] = vale.apply(lambda x: 1 if ((x['High'] > x['High-1']) &
                                                 (x['Low'] > x['Low-1']) &
                                                 (x['Price'] > x['Price-1'])) 
                                            else 0, axis=1)

vale['bearHigh'] = vale.apply(lambda x: 1 if ((x['High'] > x['High-1']) &
                                                 (x['Low'] > x['Low-1']) &
                                                 (x['Price'] < x['Price-1'])) 
                                            else 0, axis=1)

vale['bullishLow'] = vale.apply(lambda x: 1 if ((x['High'] < x['High-1']) &
                                                 (x['Low'] < x['Low-1']) &
                                                 (x['Price'] > x['Price-1'])) 
                                            else 0, axis=1)

vale['bearLow'] = vale.apply(lambda x: 1 if ((x['High'] < x['High-1']) &
                                                 (x['Low'] < x['Low-1']) &
                                                 (x['Price'] < x['Price-1'])) 
                                            else 0, axis=1)

vale['bullishHarami'] = vale.apply(lambda x: 1 if ((x['High'] < x['High-1']) &
                                                 (x['Low'] > x['Low-1']) &
                                                 (x['Price'] > x['Price-1'])) 
                                            else 0, axis=1)

vale['bearHarami'] = vale.apply(lambda x: 1 if ((x['High'] < x['High-1']) &
                                                 (x['Low'] > x['Low-1']) &
                                                 (x['Price'] < x['Price-1'])) 
                                            else 0, axis=1)
```
