```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from technical_indicators import moving_average, bollinger_bands, relative_strength_index, stochastic_oscillator

# Load data from CSV
df = pd.read_csv("../databases/stock_data.csv")

```

```python
df = df.drop('Change %', axis=1)
df.columns = ['Date', 'Close', 'Open', 'High', 'Low', 'Volume']

df.head()
```

```python
# Calculate moving average
df['MA_8'] = moving_average(df['Close'], 8)
df['MA_56'] = moving_average(df['Close'], 56)
df['MA_224'] = moving_average(df['Close'], 224)

# Calculate Bollinger Bands
df['MA_8'], df['Upper Band_8'], df['Lower Band_8'] = bollinger_bands(df['Close'], 8)
df['MA_56'], df['Upper Band_56'], df['Lower Band_56'] = bollinger_bands(df['Close'], 56)
df['MA_224'], df['Upper Band_224'], df['Lower Band_224'] = bollinger_bands(df['Close'], 224)

# Calculate Relative Strength Index
df['RSI_8'] = relative_strength_index(df['Close'], 8)
df['RSI_56'] = relative_strength_index(df['Close'], 56)
df['RSI_224'] = relative_strength_index(df['Close'], 224)

# Plot the data with technical indicators
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA_8'], label='Moving Average (8)')
plt.plot(df['MA_56'], label='Moving Average (56)')
plt.plot(df['MA_224'], label='Moving Average (224)')
plt.plot(df['Upper Band_8'], label='Upper Band (8)')
plt.plot(df['Lower Band_8'], label='Lower Band (8)')
plt.plot(df['Upper Band_56'], label='Upper Band (56)')
plt.plot(df['Lower Band_56'], label='Lower Band (56)')
plt.plot(df['Upper Band_224'], label='Upper Band (224)')
plt.plot(df['Lower Band_224'], label='Lower Band (224)')
plt.plot(df['RSI_8'], label='Relative Strength Index (8)')
plt.plot(df['RSI_56'], label='Relative Strength Index (56)')
plt.plot(df['RSI_224'], label='Relative Strength Index (224)')

plt.plot(df['Stochastic Oscillator'], label='Stochastic Oscillator')
plt.plot(df['ADX'], label='Average Directional Index')
plt.legend()
plt.show()
```
